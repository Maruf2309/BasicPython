```python
# Import pandas and numpy
import pandas as pd
import numpy as np

# Create a sample dataset of sales and conditions
data = {
    "sales": [1000, 1500, 1200, 800, 900, 2000, 1800, 1300],
    "condition": ["good", "good", "bad", "bad", "good", "good", "bad", "good"]
}

# Convert the data to a pandas dataframe
df = pd.DataFrame(data)

# Define a function to assign grades based on sales and condition
def grade_sales(sales, condition):
    # If sales are above 1500 and condition is good, assign grade A
    if sales > 1500 and condition == "good":
        return "A"
    # If sales are between 1000 and 1500 and condition is good, assign grade B
    elif 1000 <= sales <= 1500 and condition == "good":
        return "B"
    # If sales are below 1000 and condition is good, assign grade C
    elif sales < 1000 and condition == "good":
        return "C"
    # If sales are above 1500 and condition is bad, assign grade D
    elif sales > 1500 and condition == "bad":
        return "D"
    # If sales are between 1000 and 1500 and condition is bad, assign grade E
    elif 1000 <= sales <= 1500 and condition == "bad":
        return "E"
    # If sales are below 1000 and condition is bad, assign grade F
    else:
        return "F"

# Apply the function to the dataframe and create a new column for grades
df["grade"] = df.apply(lambda row: grade_sales(row["sales"], row["condition"]), axis=1)

# Print the dataframe with grades
print(df)
```

#### Outlet & Client wise grade

```python
# Import pandas and numpy
import pandas as pd
import numpy as np

# Create a sample dataset of clients, sales and regions
data = {
    "client": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "sales": [15000, 8000, 12000, 10000, 5000, 25000],
    "region": [1, 2, 1, 2, 2, 1]
}

# Convert the data to a pandas dataframe
df_1 = pd.DataFrame(data)

# Define a function to assign grades based on sales and region
def grade_clients(sales, region):
    # If region is 1, use the following criteria
    if region == 1:
        # If sales are above 10000, assign grade A
        if sales > 10000:
            return "A"
        # If sales are between 5000 and 10000, assign grade B
        elif 5000 <= sales <= 10000:
            return "B"
        # If sales are below 5000, assign grade C
        else:
            return "C"
    # If region is 2, use the following criteria
    elif region == 2:
        # If sales are above 20000, assign grade A
        if sales > 20000:
            return "A"
        # If sales are between 10000 and 20000, assign grade B
        elif 10000 <= sales <= 20000:
            return "B"
        # If sales are below 10000, assign grade C
        else:
            return "C"
    # If region is not 1 or 2, assign grade N/A
    else:
        return "N/A"

# Apply the function to the dataframe and create a new column for grades
df_1["grade"] =df_1.apply(lambda row: grade_clients(row["sales"], row["region"]), axis=1)

# Print the dataframe with grades
print(df_1)
```
