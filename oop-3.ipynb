## Methods

* No need to store seperately, method is all about creating functions
* <b> Class & method </b> (method will take the variable inside it by self)

```python
class animal:
    def get_value(self, age, food, color):
        self.age = age
        self.food = food
        self.color = color
        
    def printall(self):
        print(f"Age:{self.age}, Food:{self.food}, Color:{self.color}")
        
## Object-1
Tiger=animal()
Tiger.get_value(5, 'Meat', 'Yellow')
Tiger.printall()

## Object-2
Dog=animal()
Dog.get_value(3, 'Chicken','Black')
Dog.printall()

## Object-3
Cat=animal()
Cat.get_value(4, 'Fish', 'White')
Cat.printall()       
```

### Using Constructor will be more effective!

```python
# Constructor
class animal:
    def __init__(self, name, age, food, color):
        self.name = name
        self.age = age
        self.food = food
        self.color = color
        
    def printall(self):
        print(f" Name:{self.name}, Age:{self.age}, Food:{self.food}, Color:{self.color}")
        
## Object Define
Tiger = animal('Tiger', 10, 'Meat','Black-White') # Value passing while assing to class
Tiger.printall()

Dog = animal('Dog',2, 'Chicken','Black')       # Value passing while assing to class
Dog.printall()
 
Cat = animal('Cat', 1, 'Fishes', 'White')   # Value passing while assing to class
Cat.printall()
```

### Task1

* Adding sum of two numbers by using class, method & constructor!

```python
## Beginner Level

class Calculator:
    
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        
    def add(self):
        return self.num1 + self.num2 
    
## Object
num1 = 4
num2 = 5

calculator = Calculator(num1, num2)
result = calculator.add()
print('Sum of', num1, "and", num2, "is", result)
```

```python
class adding:
    
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def sum_number(self):
        return self.a + self.b
    
sum_of_numbers = adding(2, 3)
sum_of_numbers.sum_number()
```
