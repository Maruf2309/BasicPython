# Visualization

```python
df.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
```

- A rough graph is shown here as two, but we had 3 type of category

```python
import seaborn as sns
```

```python
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df, height=7)
```

- Additional histograph  added, which shows the max distribution/ densicity
- X axis histogram show up & y axis histogram shows rightward

```python
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df, height=7, hue='Species')
```

- Getting all species distribution

```python
sns.jointplot(x='PetalLengthCm', y='PetalWidthCm', data=df, height=7, hue='Species')  # Petal has better combination for identification
```

# Outliers

- Which package can help to identify outlier - sns

```python
sns.boxplot(x='PetalLengthCm', data=df) # No outlier found in given features
```

```python
sns.boxplot(y='SepalLengthCm', data=df) # No outlier found in given features
```

```python
# Show outlier as per category
sns.boxplot(x='Species', y='PetalLengthCm', data=df)  # having some outlier in specific category
```

```python
sns.boxplot(x='PetalLengthCm', y='PetalWidthCm', data=df)
```

```python
sns.kdeplot(data=df, x="PetalWidthCm") # most data are at 1.5, having no stronger meaning, KDA plot hue will clear it more
```

## KDA Plot

```python
sns.set(rc={'figure.figsize':(11.7,  8.27)})
sns.kdeplot(data=df, x="PetalWidthCm", hue='Species')  # KDA plot dosent take height
```
