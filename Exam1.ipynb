Question 2: Given the unsorted list:
[10, 6, 9, 2, 8, 1, 7, 3, 5, 4]
Apply bubble sort and show the state of the list after each pass. Also write the code in python. 

Bubble Sort is a simple sorting algorithm that repeatedly compares adjacent elements and swaps them if they are in the wrong order. Let’s apply Bubble Sort to the given list and show the state of the list after each pass.

```python
def bubbleSort(arr):
    n = len(arr)
    for i in range(n):
        # Last i elements are already in place, so the inner loop can skip them
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap the elements if they are out of order
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Given unsorted list
my_list = [10, 6, 9, 2, 8, 1, 7, 3, 5, 4]

# Apply Bubble Sort
bubbleSort(my_list)

# Print the sorted list after each pass
for i in range(len(my_list)):
    print(my_list[:i + 1])
```

Question 3: Given the unsorted list:
•	[10, 6, 9, 2, 8, 1, 7, 3, 5, 4]
Apply Selection sort and show the state of the list after each pass.


```python
def selection_sort(lst):
    n = len(lst)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if lst[j] < lst[min_index]:
                min_index = j
        # Swap the elements to sort the array
        lst[i], lst[min_index] = lst[min_index], lst[i]

# Given unsorted list
my_list = [10, 6, 9, 2, 8, 1, 7, 3, 5, 4]

# Apply Selection Sort
selection_sort(my_list)

# Print the list after each pass
for i in range(len(my_list)):
    print(my_list[:i + 1])
```

Question 4: Given the unsorted list:
•	[10, 6, 9, 2, 8, 1, 7, 3, 5, 4]
Apply Merge sort and show the state of the list after each pass.



```python
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
    L = [0] * n1
    R = [0] * n2

    for i in range(n1):
        L[i] = arr[l + i]
    for j in range(n2):
        R[j] = arr[m + 1 + j]

    i = 0
    j = 0
    k = l

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def mergeSort(arr, l, r):
    if l < r:
        m = l + (r - l) // 2
        mergeSort(arr, l, m)
        mergeSort(arr, m + 1, r)
        merge(arr, l, m, r)

# Given unsorted list
my_list = [10, 6, 9, 2, 8, 1, 7, 3, 5, 4]

# Apply Merge Sort
mergeSort(my_list, 0, len(my_list) - 1)

# Print the list after each pass
for i in range(len(my_list)):
    print(my_list[:i + 1])
```
