## Data Distribution

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')
```

```python
import pandas as pd
weight_data = [ ['tom',3.0], ['pumpkin', 3.51], ['bonk', 4.2], ['thunder', 5.50], ['oreo', 4.73], ['nya', 5.00], ['kitkat',4.55], ['bubbles', 4.90], ['sparkle',6.29], ['pebbles', 3.72] ]
df_weight = pd.DataFrame(weight_data, columns = ['Name', 'Weight(Kg)'])

df_weight
```

```python
import seaborn as sns
sns.histplot(data = df_weight, bins = 5, x='Weight(Kg)') # bins/ bucket - 5
plt.show()
```

```python
import seaborn as sns
sns.histplot(data = df_weight, bins = 6, x='Weight(Kg)') # bins/ bucket - 5
plt.show()
```

```python
import seaborn as sns
sns.histplot(data = df_weight, bins = 'auto', x='Weight(Kg)') # bins: auto, python will take the best fit
plt.show()
```

1. Most of the students tends to get Moderate level number / height pick (bar)- Normal Distribution
2. Normal Distribution: Mean & Median Value wil be qual to or close to equal or fraction difference
3. Student have preperation to get 70% of marks but actually got 90% which is called <b> Negative Skewed </b> where Mean < Median
4. Student have preperation to get 70% of marks but actually got 50% which is called <b> Positive Skewed </b> where Mean > Median
